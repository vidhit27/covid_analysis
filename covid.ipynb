```python
#Dropping the columns which are not required
covid.drop(['Time','Sno','ConfirmedIndianNational','ConfirmedForeignNational'], inplace=True, axis=1)
```

```python
covid.head()
```

```python
covid['Active_Cases']=covid['Confirmed']-(covid['Cured']+covid['Deaths'])
```

```python
covid.head()
```

```python
covid.dropna(inplace=True)
```

```python
covid.tail()
```

```python
statewise=pd.pivot_table(covid,values=['Cured','Confirmed','Deaths'], index=['State/UnionTerritory'],aggfunc=max)
```

```python
statewise['Recovery_rate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality_rate']=statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise=statewise.sort_values(by='Confirmed',ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
#Top 10 states with active cases
top_cases=covid.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by='Active_Cases', ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
plt.title('Top 10 States with Active Cases')
ax=sns.barplot(data=top_cases.iloc[:10],x='State/UnionTerritory', y='Active_Cases',linewidth=2,edgecolor='black')
```

So Maharashtra has the most number of Active Cases in India

```python
#Top 10 states with most deaths
top_deaths=covid.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by='Deaths', ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
plt.title('Top 10 States with Most Deaths')
ax=sns.barplot(data=top_deaths.iloc[:10],x='State/UnionTerritory', y='Deaths',linewidth=2,edgecolor='black')
```

So, Maharashtra has reported the most number of deaths in the country

```python
covid['Date']=pd.to_datetime(covid['Date'],format='%Y-%m-%d')
fig=plt.figure(figsize=(16,9))
ax=sns.lineplot(data=covid[covid['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Delhi','Tamil Nadu','Uttar Pradesh'])],x='Date',y='Active_Cases',hue='State/UnionTerritory')
```

```python
vaccine.rename(columns={'Updated On':'Vaccine_Date'},inplace=True)
```

```python
vaccine.drop(columns=['Sputnik V (Doses Administered)','AEFI','18-44 Years (Doses Administered)','45-60 Years (Doses Administered)','60+ Years (Doses Administered)'],inplace=True)
```

```python
#Male vs Female Vaccination
# male=vaccine['Male(Individuals Vaccinated)'].sum()
# female=vaccine['Female(Individuals Vaccinated)'].sum()
# px.pie(names=['Male','Female'], values=['male','female'], title='Male & Female Vaccination')
male=np.sum(vaccine['Male(Individuals Vaccinated)'])
female=np.sum(vaccine['Female(Individuals Vaccinated)'])

gen=[male,female]
plt.title('Vaccination Distribution (Gender)')
plt.pie(gen,labels=['Male','Female'],autopct="%1.1f%%",colors=['lightgreen','lightblue'],shadow=True,explode=(0.1,0))
plt.show()
```

```python
#Some states are miswritten as India so removing those states
vaccine=vaccine[vaccine['State']!='India']
```

```python
vaccine.head()
```

```python
#Most Vaccinated States
vaccine.rename(columns={'Total Individuals Vaccinated':'Total'},inplace=True)
max_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values(by='Total', ascending=False)[:5]
```

```python

fig=plt.figure(figsize=(10,5))
plt.title('Top 5 Vaccinated States')
x=sns.barplot(data=max_vac.iloc[:5],y=max_vac.Total,x=max_vac.index,linewidth=2,edgecolor='black')
plt.xlabel('States')
plt.ylabel('Total Vaccination')
plt.show()
```

```python
#Least Vaccinated States
min_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac=max_vac.sort_values(by='Total', ascending=True)[:5]
min_vac
```

```python
fig=plt.figure(figsize=(16,5))
plt.title('Bottom 5 Vaccinated States')
x=sns.barplot(data=min_vac.iloc[:5],y=min_vac.Total,x=min_vac.index,linewidth=2,edgecolor='black')
plt.xlabel('States')
plt.ylabel('Total Vaccination')
plt.show()
```
